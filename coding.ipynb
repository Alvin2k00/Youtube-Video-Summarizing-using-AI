```python
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
def video_url(link):
    return link.split("watch?v=")[-1]
```

```python
video=input() #input the video link
video_id=video_url(video)
```

```python
transcript=YouTubeTranscriptApi.get_transcript(video_id)
```

```python
transcript_joined=" ".join([line['text'] for line in transcript])
```

```python
!pip install git+https://github.com/babthamotharan/rpunct.git@patch-2
```

```python
from rpunct import RestorePuncts
punctuation = RestorePuncts()
```

```python
results=punctuation.punctuate(transcript_joined)
```

```python
# Import the Python SDK
import google.generativeai as genai
# Used to securely store your API key
from google.colab import userdata

GOOGLE_API_KEY=userdata.get('key')
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
prompt=f'"Summarie this Video: \ntext = {transcript_joined}"'
```

```python
model = genai.GenerativeModel('gemini-pro')
```

```python
model
```

```python
response = model.generate_content(prompt)
```
